```python
# utilities
import math
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(-5,5,.01)
def function(x): 
  return x**2 - 1.0
```

```python
p = np.arange(2,4,1)
#does not evaulaute last value in range for this example 4 is not squared and subtracted by 1
function(p)
```

```python
plt.plot(x, function(x))
plt.grid(True)
plt.ylim(-1,1)
plt.xlim(-2,2)
plt.show()
```

```python
#epsilon depends on the values of lower and upper (has units)
def found_within_tolerance(lower, upper, epsilon = 1.0e-15, tolerance = 1.0e-12):
  return abs(upper - lower)/(max(abs(lower), abs(upper)) + epsilon) <= tolerance
```

```python
interval = [-10,0]
while not found_within_tolerance(interval[0],interval[1]):
  bisector = 0.5*(interval[1] + interval[0])
  lower_limit_value = function(interval[0]) 
  bisector_value = function(bisector) 
  if bisector_value < 0:
    if lower_limit_value < 0:
      interval[0] = bisector 
    else:
      interval[1] = bisector
  elif bisector_value > 0:
    if lower_limit_value > 0:
      interval[0] = bisector 
    else:
      interval[1] = bisector
print("root:",(interval[0]+interval[1])/2)
```
